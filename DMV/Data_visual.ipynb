```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

# Import the "City_Air_Quality.csv" dataset.

```python
df=pd.read_csv("city_day.csv")
```

```python
df
```

```python
delphi_df = df[df['City'] == 'Delhi']
```

```python
mean_value = delphi_df['PM2.5'].mean()
delphi_df['PM2.5'].fillna(value = mean_value, inplace = True)
```

#  Identify the relevant variables for visualizing AQI trends, such as date, pollutant levels,
and AQI values.

```python
date_column = "Date"
aqi_column = "AQI"
pollutants = ["PM2.5", "PM10", "CO"]
```

#  Create line plots or time series plots to visualize the overall AQI trend over time

```python
plt.figure(figsize=(12, 6))
plt.plot(df[date_column], df[aqi_column], label="AQI", color='b')
plt.xlabel("Date")
plt.ylabel("AQI")
plt.title("AQI Trend Over Time")
plt.legend()
plt.show()
```

#  Plot individual pollutant levels

```python
plt.figure(figsize=(12, 6))
for pollutant in pollutants:
    plt.plot(df[date_column], df[pollutant], label=pollutant)
plt.xlabel("Date")
plt.ylabel("Pollutant Level")
plt.title("Pollutant Trends Over Time")
plt.legend()
plt.show()
```

# Use bar plots or stacked bar plots to compare the AQI values across different dates or 
time periods.

```python
plt.figure(figsize=(12, 6))
plt.bar(df[date_column], df[aqi_column], label="AQI")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.title("AQI Comparison Across Dates")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

# Use scatter plots or bubble charts to explore the relationship between AQI values and 
pollutant levels.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df["PM2.5"], df[aqi_column], label="PM2.5", alpha=0.5)
plt.scatter(df["PM10"], df[aqi_column], label="PM10", alpha=0.5)
plt.scatter(df["CO"], df[aqi_column], label="CO", alpha=0.5)
plt.xlabel("Pollutant Levels")
plt.ylabel("AQI")
plt.title("Scatter Plot of AQI vs Pollutant Levels")
plt.legend()
plt.show()
```
